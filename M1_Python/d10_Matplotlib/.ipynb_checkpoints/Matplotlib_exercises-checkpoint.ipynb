### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
#print(data.head())
plt.plot(data['month_number'], data['total_profit'] )
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')

```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here

fig_e2 = plt.figure()
ax_e2 = fig_e2.add_subplot(1, 1, 1)

ax_e2.plot(data['month_number'], data['total_profit'], marker='o', mec='r', color='r', linestyle='--',
          linewidth=3, mfc='k', label='profit data of last year')
ax_e2.set_title('Company profit per month')
ax_e2.set_xlabel('Month number')
ax_e2.set_yticks([100000, 200000, 300000, 400000, 500000])
ax_e2.set_ylabel('Total profit')
ax_e2.set_xticks(np.arange(1,13))
ax_e2.legend(loc='lower right')
```

### Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

The graph should look like this:


```python
data.head(12)
```

```python
# Write your code here
fig, ax_e3 = plt.subplots(figsize=(10,8))
ax_e3.plot(data['month_number'], data['facecream'], markersize=10,linewidth=4,marker='o', color='b',label='facecream')
ax_e3.plot(data['month_number'], data['facewash'], markersize=10, linewidth=4, marker='o',color='tab:orange',label='facewash')
ax_e3.plot(data['month_number'], data['toothpaste'], markersize=10,linewidth=4, marker='o',color='g',label='toothpaste')
ax_e3.plot(data['month_number'], data['bathingsoap'],markersize=10,linewidth=4, marker='o',color='r', label='bathingsoap')
ax_e3.plot(data['month_number'], data['shampoo'], markersize=10,linewidth=4, marker='o',color='tab:purple', label='shampoo')
ax_e3.plot(data['month_number'], data['moisturizer'],markersize=10, linewidth=4, marker='o',color='tab:brown',label='moisturizer')
ax_e3.set_xlabel('Month number')
ax_e3.set_yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
ax_e3.set_ylabel('Sales units in numbers')
ax_e3.set_xticks(np.arange(1,13))
ax_e3.legend()
```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
# Write your code here
#print(data.head())
fig_e4, axe_e4 = plt.subplots(figsize=(10,6))
axe_e4.grid(b=True,linestyle='--',linewidth=2)
axe_e4.set_xticks(np.arange(1,13))
axe_e4.tick_params(axis='x', labelsize="x-large" )
axe_e4.tick_params(axis='y', labelsize="x-large" )
axe_e4.set_ylabel("Number units sold", fontsize="xx-large")
axe_e4.set_xlabel("Month Number", fontsize="xx-large")
axe_e4.scatter(data['month_number'], data['toothpaste'], label="Toothpaste", s=150)
axe_e4.legend(loc="upper left", fontsize="x-large")

```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
# Write your code here

width=0.35
fig_e5, ax_e5 = plt.subplots(figsize=(10,6))
ax_e5.grid(linestyle="--", linewidth=2)
ax_e5.set_xticks(np.arange(0,12))
ax_e5.set_xticklabels(np.arange(1,13),fontsize="xx-large")
#ax_e5.set_yticks(np.arange(0,8))
ax_e5.set_yticklabels(np.arange(0, 4000, 500), fontsize="xx-large")
ax_e5.set_title("Facecream Vs Facewash units sold", fontsize="xx-large")
ax_e5.set_ylabel("Units Sold", fontsize="xx-large")
ax_e5.set_xlabel("Month Number", fontsize="xx-large")
ax_e5.bar([i - width/2 for i in range(len(data["month_number"]))], data["facecream"], width, label='Facecream')
ax_e5.bar([i + width/2 for i in range(len(data["month_number"]))], data["facewash"], width, label='Facewash')
display(data.head())
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
fig_e8, ax_e8 = plt.subplots(figsize=(10,6))
ax_e8.set_title("Sales Data", fontsize="xx-large")
catch_data =["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
totals = [data[i].sum() for i in catch_data]
ax_e8.pie(totals, autopct = '%0.2f%%', labels=catch_data)
ax_e8.legend(catch_data, loc="lower right")
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
# Write your code here
fig_e9, ax_e9 = plt.subplots(2,1) # Subplot arguments (number of rows, number of columns)
                                    # if a layour representing a grid is given then you index 
                                      # the subplot like a numpy matrix.
# First plot
ax_e9[0].plot(data['month_number'], data['bathingsoap'], 'ko-')
ax_e9[0].set_title('Sales data of bathing soap')
ax_e9[0].tick_params(labelbottom=False)

# Second plot
ax_e9[1].plot(data['month_number'], data['facewash'],'ro-')
ax_e9[1].set_title('Sales data of facewash')
ax_e9[1].set_xticks(np.arange(1,13))
ax_e9[1].set_xlabel("Month number")
ax_e9[1].set_ylabel("sales units")
display(data.head())
```

###Â Exercise 10: Read all product sales data and show it using the stack plot.
The Stack plot should look like this:


```python
# Write your code here
fig_e10, ax_10 = plt.subplots(figsize=(10,6))
ax_10.stackplot(data["month_number"], [data['facecream'], data['facewash'],data['toothpaste'],\
                                       data['bathingsoap'], data['shampoo'], data['moisturizer']],\
                labels=['facecream','facewash','toothpaste', 'bathingsoap', 'shampoo','moisturizer'])
ax_10.legend(fontsize="xx-large", loc="upper left")
display(data.head())
```
