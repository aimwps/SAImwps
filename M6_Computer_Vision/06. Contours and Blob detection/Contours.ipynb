# Computer Vision Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Contours

```python
# Have a look at the image below
color_coins = cv2.imread('img/coins5.jpg', cv2.IMREAD_COLOR)
rgb_coins = cv2.cvtColor(color_coins,cv2.COLOR_BGR2RGB)

plt.figure(figsize = (12,8))
plt.imshow(rgb_coins);
```

# GRAYSCALE & BLUR
1. Change to grayscale
2. Invert colours with binary threshold & apply a gaussian blur
3. find contours

```python
# Convert the image above to grayscale
gray_coins = cv2.cvtColor(color_coins.copy(), cv2.COLOR_BGR2GRAY)
plt.imshow(gray_coins, cmap='gray')
```

```python
_, inv_gray_coins = cv2.threshold(gray_coins,220, 255, cv2.THRESH_BINARY_INV)
gb3_coins = cv2.GaussianBlur(inv_gray_coins, (15,15),0)
plt.imshow(gb3_coins, cmap='gray')
```

```python
gcontours, hg = cv2.findContours(gb3_coins.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print(len(gcontours))
gcoins_n_contours = color_coins.copy()
cv2.drawContours(gcoins_n_contours, gcontours, -1, (0,255,0),2)
plt.imshow(gcoins_n_contours)
```

# USING EDGE DETECTION(CANNY)
1. Apply Canny

```python
# Apply Canny to find edges and display the image 
line_coins = cv2.Canny(gray_coins.copy(), 100, 700)
plt.imshow(line_coins, cmap='gray')
```

```python
lcontours, hl = cv2.findContours(line_coins.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
lcoins_n_contours = color_coins.copy()
cv2.drawContours(lcoins_n_contours, lcontours, -1, (0,255,0),2)
plt.imshow(lcoins_n_contours)
```

```python
sorted_contours = sorted(lcontours, key=cv2.contourArea, reverse=True)
print(len(sorted_contours))
scoins_n_contours = color_coins.copy()
cv2.drawContours(scoins_n_contours, sorted_contours[0:8], -1, (0,255,0),2)
plt.imshow(scoins_n_contours)
```

## Sorting Contours

### By area

```python
def get_contour_areas(contours):
    all_areas = []
    for contour in contours:
        contour_area = cv2.contourArea(contour)
        all_areas.append(contour_area)
    return all_areas
```

```python
# We can sort contours with any function, a really useful way of sorting is by area, in many situations we might
# only want to find the biggest or smallest contours, I'm providing the function I use to get all contour areas
# This function is useful to be able to see at a glance all contours areas and get a sense of what is the biggest 
# And which one is the smallest one, however we will not use this to sort them we can just pass cv2.contourArea 
# To the function that is going to be sorting it later
# Call get_contour_areas an pass in the list of contours you got with openCV and then store the results in a var
get_contour_areas(lcontours)
```

### By X coordinate (left to right)

```python
def get_contour_x_center_coordinate(contours):    
    if cv2.contourArea(contours) > 10:
        M = cv2.moments(contours)
        return(int(M['m10']/M['m00']))
```

[OpenCV Moments](https://www.learnopencv.com/find-center-of-blob-centroid-using-opencv-cpp-python/)

```python
# This is an useful function you can use when you need to sort contours left to right
# Sort the contours this time left to right
l2r_contours = sorted(lcontours, key=get_contour_x_center_coordinate, reverse=False)
for contour in l2r_contours:
    l2r_coins = color_coins.copy()
    cv2.drawContours(l2r_coins, [contour], -1, (0,255,0), 5 )
    plt.imshow(l2r_coins)
    
```
