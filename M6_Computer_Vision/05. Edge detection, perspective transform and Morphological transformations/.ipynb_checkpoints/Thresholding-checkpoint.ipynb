# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Have a look at the 2 images below
text = cv2.imread('img/t1.png', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray')
```

```python
text_2 = cv2.imread('img/t2.png')
plt.figure(figsize = (12,10))
plt.imshow(text_2, cmap='gray')
```

```python
# Apply threshold to the images above to make them look better, try to find the best one
_, thresh_binary = cv2.threshold(text, 128, 255, cv2.THRESH_BINARY)
_, thresh_trunc = cv2.threshold(text, 128, 255, cv2.THRESH_TRUNC)
_, thresh_tozero = cv2.threshold(text, 128, 255, cv2.THRESH_TOZERO)
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray')
plt.figure(figsize = (12,10))
plt.imshow(thresh_binary, cmap='gray')
plt.figure(figsize = (12,10))
plt.imshow(thresh_trunc, cmap='gray')
plt.figure(figsize = (12,10))
plt.imshow(thresh_tozero, cmap='gray')
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C
adaptive_mean_c = cv2.adaptiveThreshold(text,255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 10)
adaptive_gaussian = cv2.adaptiveThreshold(text,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 8)
plt.figure(figsize = (12,10))
plt.imshow(adaptive_mean_c, cmap='gray')
plt.figure(figsize = (12,10))
plt.imshow(adaptive_gaussian, cmap='gray')
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image
_, otsu = cv2.threshold(text, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.figure(figsize = (12,10))
plt.imshow(otsu, cmap='gray')
```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding
gb = cv2.GaussianBlur(text, (5,5),0)
_, otsu_gb = cv2.threshold(gb, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.figure(figsize = (12,10))
plt.imshow(gb, cmap='gray')
plt.figure(figsize = (12,10))
plt.imshow(otsu_gb, cmap='gray')
```
